```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import preprocessing
%matplotlib inline
```

```python
a = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv ",sep=',')
```

```python
b = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv",sep=',')
```

```python
c=[a,b]
d=pd.concat(c)
```

```python
d.head()
```

```python
d.shape
```

```python
e = d.dropna(subset=["Survived"])
```

```python
e
```

```python
e.isnull().sum()
```

```python
e.insert(loc =2,column = "Has_cabin",value=e["Cabin"].notna())
```

```python
e.head()
```

```python
import re
```

```python
e["Title"]=e.Name.apply(lambda x : re.search('([A-Z][a-z]+)\.',x).group(1))
e["Title"].unique()
```

```python
list = ["Don","Rev","Dr","Mme","Ms","Major","Lady","Sir","Mlle","Col","Capt","Countess","Jonkheer"]
e["Title"] = e["Title"].replace(list,"Special")
```

```python
e["Title"].unique()
```

```python
e
```

```python
e =e.fillna(e.mean())
```

```python
e.mode()
```

```python
e.isnull().sum()
```

```python
e["Categorical_age"] = pd.qcut(e.Age,q= 4,labels = False)
```

```python
e["Categorical_fare"] =  pd.qcut(e.Fare,q=4,labels =False)
```

```python
e.head()
```

```python
e=  e.drop(["Cabin", "Name", "PassengerId", "Ticket", "Age", "Fare", "SibSp", "Parch" ],axis = 1)
```

```python
e.head()
```

```python
e = pd.get_dummies(e)
```

```python
e.head()
```

```python
X = e.drop(["Survived"],axis = 1)
```

```python
y = e["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
reg.fit(X_train,y_train)
```

```python
pred = reg.predict(X_test)
```

```python
result = pd.DataFrame({"Predicted":pred,"Actual":y_test})
```

```python
result
```

```python
reg.score(X_test,y_test)
```
