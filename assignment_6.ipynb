```python
df= pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
df = pd.DataFrame(df)
```

```python
plt.plot(df.Date,df.MaxTemp)
plt.xlabel("Date")
plt.ylabel("maxTemp")
```

```python
plt.scatter(df.Date,df.MaxTemp)
plt.xlabel("Date")
plt.ylabel("maxTemp")
```
