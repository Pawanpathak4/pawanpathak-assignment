```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stopword = set(stopwords.words('english'))
```

```python
a['texts'] = a['texts'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopword)]))
```

```python
from sklearn.feature_extraction import text
```

```python
b = text.TfidfVectorizer()
```

```python
x = b.fit_transform(a["texts"])
```

```python
x.shape
```

```python
b.vocabulary_
```

```python
a
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
d= pd.DataFrame(a[0])
```

```python
a2 = SelectKBest(chi2,k=4).fit_transform(x,d)
```

```python
a2.shape
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
shuffle = StratifiedShuffleSplit(n_splits=101,random_state=7)
```

```python
d= d.values.reshape(18086)
```

```python
for train_index, test_index in shuffle.split(a2, d):
    x_train, x_test = a2[train_index], a2[test_index]
    d_train, d_test = d[train_index], d[test_index]
```

## vooting classifier

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```
