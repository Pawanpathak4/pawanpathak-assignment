```python
from sklearn.datasets import load_digits
import pandas as pd
import numpy as np
```

```python
data = load_digits()
X = data.data
Y= data.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 8)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters= 10,max_iter= 1000,random_state=7).fit(X_train)
```

```python
pred = model.predict(X_test)
```

```python
pre = pd.DataFrame(pred,columns=["Value"])
```

```python
pre.isnull().sum()
```

```python
pre["y"] = Y_test
```

```python
pre.head()
```

```python
labels = np.zeros_like(Y_test)
```

```python
from scipy import stats
for i in range(10):
    m =(pred == i)
    labels[m] = stats.mode(Y_test[m])[0]
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(labels,Y_test)
```

```python
print(metrics.confusion_matrix(labels,Y_test))
```
